```python
import math
import graphviz
```

```python
instance = "instance_20.dat"
location = "./trabOtimizacao/instances/" + instance
data = open(location).read().split()

sol, lista = int(data[0]), (data[1:])
length=len(lista)
sol,lista
```

```python
class Graph:
    def __init__(self) -> None:
        self.V = [] #list of vertices
        self.E = [] #list of pairs
    def __repr__(self):
        return (f"Vertices: {self.V} \nEdges: {self.E}")

    def addEdge(self,u:int,v:int):
        self.E.append((u,v))
    def addVertex(self,v:int):
        self.V.append(v)

    def toGraphviz(self):
        gv = ["graph {"]
        for v in self.V:
            gv.append(f"  {v};")
        for u, v in self.E:
            gv.append(f"  {u} -- {v};")
        gv.append("}")
        return "\n".join(gv)
    
def isPerfectSquare(n): 
    if math.ceil(math.sqrt(n)) == math.floor(math.sqrt(n)):
        return True
    else:
        return False

def get_ps_from_list(data):
    """
    input: lista de números
    return: lista de tuplas contendo as duplas que cuja soma
    é um quadrado perfeito
    """
    ps = []
    qtt = {}
    for i in range(len(data)):
        number1 = int(data[i])
        qtt[number1] = qtt.get(number1, 0) + 1
        for j in range(i+1,len(data)):
            number2 = int(data[j])
            if (isPerfectSquare(number1+number2)):
                ps.append((number1,number2))
    return ps,qtt
        
    
def makePerfectSquaresGraph(data,instance_length):
    graph = Graph()
    for i in range(instance_length):
        number1 = int(i)
        graph.addVertex(number1)
        for j in range(i+1,instance_length):
            number2 = int(data[j])
            if (isPerfectSquare(number1+number2)):
                graph.addEdge(i,j) 
    return graph
```

```python
ps, qtt = get_ps_from_list(lista)
ps,qtt
```

```python
graph = makePerfectSquaresGraph(lista,length)
print(graph)
graphviz.Source(graph.toGraphviz())
```
